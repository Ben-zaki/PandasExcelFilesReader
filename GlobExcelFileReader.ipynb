#### reading multiple excel workbooks

```python
#selecting all files with xls extension only 
path = "GlobDataTest\*.xls"
```

```python
excelfiles = glob.glob(path)
```

```python
excelfiles
```

```python
type(excelfiles)
```

```python
df0 = pd.read_excel(excelfiles[0])
df0.head(3)
```

```python
#create an empty dataframe
dfglobal = pd.DataFrame()
```

```python
for file in excelfiles:
    df = pd.read_excel(file)
    dfglobal = pd.concat([dfglobal,df])
```

```python
dfglobal.head()
```

```python
dfglobal.to_excel('CombinedData.xlsx')
```
