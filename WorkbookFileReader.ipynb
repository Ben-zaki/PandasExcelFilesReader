#### reading multiple sheets from a workbook


```python
path = "Workbook\MultipleSheets.xlsx"
```

```python
# outputs a dictionary
df = pd.read_excel(path,sheet_name = [0,1,2,3,4])
```

```python
type(df)
```

```python
df[0].head()
```

```python
CombinedSheets = pd.concat([df[0],df[1],df[2],df[3],df[4]])
```

```python
type(CombinedSheets)
```

```python
CombinedSheets.head()
```

```python
CombinedSheets.to_excel('ConcatenatedSheets.xlsx')
```
